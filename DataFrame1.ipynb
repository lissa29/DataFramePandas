```python
import pandas as pd
df = pd.read_csv("estudiante.csv")
df
```

```python
df1 = pd.read_csv("asignatura.csv")
df1
```

```python
df2 = pd.read_csv("nivel.csv")
df2
```

```python
file_name = "rendimiento.json"
dfh = pd.read_json(file_name)
dfh
```

```python
join_df = pd.merge(df, df2, on='id_estudiante', how='outer')
join_df
```

```python
all_datadf = pd.merge(join_df, dfh, on='id_estudiante', how='outer')
all_datadf
```

```python
all_datadf[(all_datadf.promedio>80) & (all_datadf.promedio<90)] #estudiantes con promedio muy bueno: promedio mayor a 80 y menor a 90
```

```python
import numpy as np
all_datadf.groupby(['facultad']).agg({'promedio': np.mean}) #avg del promedio de notas de estudiantes por facultad
```

```python
all_datadf.groupby(['facultad']).agg({'facultad':'count'} )
```

```python
all_datadf.groupby(['facultad']).agg({'promedio': np.max })#promedio mas alto por facultad
```
